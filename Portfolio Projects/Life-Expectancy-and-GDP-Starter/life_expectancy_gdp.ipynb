```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

all_data = pd.read_csv("all_data.csv")
print(all_data.head())
print(all_data.describe(include="all"))
print(all_data.dtypes)
print(len(all_data))
print(all_data.columns)
all_data.rename(columns = {
  "Life expectancy at birth (years)":"life_expectancy"},inplace=True)
print(all_data.columns)
all_data.replace("United States of America","USA")

sns.countplot(all_data.Country)
plt.show()
plt.clf()

countries = all_data.Country.unique()
sns.boxplot(data=all_data, x="Country", y="life_expectancy")
plt.xticks(rotation=30)
plt.show()
plt.clf()
```
