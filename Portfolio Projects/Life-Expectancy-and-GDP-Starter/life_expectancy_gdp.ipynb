```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

all_data = pd.read_csv("all_data.csv")
print(all_data.head())
print(all_data.describe(include="all"))
print(all_data.dtypes)
print(len(all_data))
print(all_data.columns)
all_data.rename(columns = {
  "Life expectancy at birth (years)":"life_expectancy"},inplace=True)
print(all_data.columns)
all_data.Country = all_data.Country.replace("United States of America","USA")

sns.countplot(all_data.Country)
plt.show()
plt.clf()

sns.boxplot(data=all_data, x="Country", y="life_expectancy")
plt.xticks(rotation=30)
plt.show()
plt.clf()

plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
sns.boxplot(data=all_data, x="Country", y="life_expectancy")
plt.xticks(rotation=30)

plt.subplot(1,2,2)
sns.boxplot(data=all_data,x="Country",y="GDP")
plt.xticks(rotation=30)

plt.subplots_adjust(wspace=0.4)

plt.show()
plt.clf()

plt.figure(figsize=(12,10))
i = 1
countries = all_data.Country.unique()
for country in countries:
    plt.subplot(2,3,i)
    i += 1
    plt.scatter(all_data.Year[all_data.Country == country],all_data.life_expectancy[all_data.Country == country])
    plt.xlabel("Year")
    plt.ylabel("Life Expectancy at Birth [yr]")
    plt.title(country)
    if country == "Zimbabwe":
        plt.axis([1998,2017,40,82])
    else:
        plt.axis([1998,2017,70,82])

plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.show()
plt.clf()

plt.figure(figsize=(12,10))
i = 1
for country in countries:
    plt.subplot(2,3,i)
    i += 1
    plt.scatter(all_data.Year[all_data.Country == country],all_data.GDP[all_data.Country == country])
    plt.xlabel("Year")
    plt.ylabel("Life Expectancy at Birth [yr]")
    plt.title(country)
    #if country == "Zimbabwe":
        #plt.axis([1998,2017,40,82])
    #else:
        #plt.axis([1998,2017,70,82])

plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.show()
plt.clf()

plt.figure(figsize=(12,10))
i = 1
for country in countries:
    plt.subplot(2,3,i)
    i += 1
    plt.scatter(all_data.GDP[all_data.Country == country],all_data.life_expectancy[all_data.Country == country])
    plt.xlabel("GDP")
    plt.ylabel("Life Expectancy at Birth [yr]")
    plt.title(country)
    #if country == "Zimbabwe":
        #plt.axis([1998,2017,40,82])
    #else:
        #plt.axis([1998,2017,70,82])

plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.show()
plt.clf()
```
